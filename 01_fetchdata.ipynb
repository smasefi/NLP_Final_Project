## Scrape and Format

```python
import re
from urllib.parse import urlparse

def get_book_id(url):
    path = urlparse(url).path.lower()

    if "luxemburg" in path and "reform-revolution" in path: # democratic socialism
        return "luxemburg_reform-or-revolution"
    elif "mao" in path and "guerrilla-warfare" in path:
        return "mao_on-guerrilla-warfare"
    elif "mao" in path and "mswv1_5" in path:
        return "mao_correcting-mistaken-ideas"
    elif "mao" in path and "mswv1_11" in path:
        return "mao_tactics-against-japanese-imperialism"
    elif "mao" in path and "mswv1_16" in path:
        return "mao_on-practice"
    elif "mao" in path and "mswv1_15" in path:
        return "mao_win-the-masses"
    elif "mao" in path and ("guerrilla-warfare" in path):
        return "mao_on-guerrilla-warfare"
    elif "marx" in path and "communist-manifesto" in path:
        return "marx_communist-manifesto"
    elif "fourier" in path and "/works/" in path:
        return "fourier_selections"
    elif "marx" in path and "wage-labour" in path:
        return "marx_wage-labour-and-capital"
    else:
        return "unknown_book"
```

```python
def get_chapter_filename(url):
    filename = url.rstrip("/").split("/")[-1]

    if filename == "intro.htm":
        return "ch00_intro.txt"

    m = re.match(r"(?:ch|chapter)(\d+)\.htm?$", filename, flags=re.I)
    if m:
        return f"ch{int(m.group(1)):02d}.txt"

    stem = re.sub(r"\.html?$", "", filename, flags=re.I)
    return f"{stem}.txt"
```

```python
from pathlib import Path

BASE_DIR = Path("corpus")
BOOKS_DIR = BASE_DIR / "books"
BOOKS_DIR.mkdir(parents=True, exist_ok=True)
```

```python
import time
import requests
from bs4 import BeautifulSoup

HEADERS = {"User-Agent": "Mozilla/5.0"}

def clean_text(text):
    return re.sub(r"\s+", " ", text or "").strip()

def extract_page(url):
    r = requests.get(url, headers=HEADERS, timeout=30)
    r.raise_for_status()
    soup = BeautifulSoup(r.text, "html.parser")

    title = clean_text(soup.title.get_text()) if soup.title else ""
    h4 = soup.find("h4")
    chapter_title = clean_text(h4.get_text()) if h4 else ""

    paragraphs = []
    for tag in soup.find_all(["p", "para"]):
        txt = clean_text(tag.get_text())
        if txt:
            paragraphs.append(txt)

    return {
        "page_title": title,
        "chapter_title": chapter_title,
        "paragraphs": paragraphs,
    }

def format_txt(url, data):
    lines = []
    lines.append("<document>")
    lines.append(f"<source_url>{url}</source_url>")
    lines.append(f"<page_title>{data['page_title']}</page_title>")
    lines.append(f"<chapter_title>{data['chapter_title']}</chapter_title>")
    lines.append("")

    for i, para in enumerate(data["paragraphs"], start=1):
        lines.append(f'<para id="{i}">{para}</para>')

    lines.append("</document>")
    return "\n".join(lines)

for url in URLS:
    book_id = get_book_id(url)
    filename = get_chapter_filename(url)

    out_dir = BOOKS_DIR / book_id
    out_dir.mkdir(parents=True, exist_ok=True)
    
    data = extract_page(url)
    text_output = format_txt(url, data)

    out_path = out_dir / filename
    out_path.write_text(text_output, encoding="utf-8")

    print(f"saved -> {out_path}")
    time.sleep(1)
```
